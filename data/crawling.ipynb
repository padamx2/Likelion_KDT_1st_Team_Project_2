```python
import bs4
import time
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import re
```

```python
link = 'https://www.10000recipe.com/issue/view.html?cid=gdubu33&types=issue&page='
page = 1
addr = "https://www.10000recipe.com"
```

```python
num = 0
recipe = {}
for i in range(1,2): # 페이지
    my_driver = webdriver.Chrome('chromedriver.exe') 
    my_driver.get(link+str(i))
    my_html = my_driver.page_source                      # 사이트 내용을 가져온다.
    my_soup = bs4.BeautifulSoup(my_html, 'html.parser')     # BeautifulSoup 객체 반환. 'html.parser' 사용.
    find = my_soup.find_all("a", class_='thumbnail')
    for f in find:
        addr_ = addr+f["href"]
        my_driver.get(addr_)
        my_html = my_driver.page_source
        my_soup = bs4.BeautifulSoup(my_html, 'html.parser')
        ma = my_soup.select('div > div > ul > a > li')
        ti = my_soup.select('div > div > div > h3')
        num+=1
        recipe["recipe"+str(num)] = {"title": "","material": []}
        for a in ma:
            recipe['recipe'+str(num)]["material"].append(re.findall("[가-핳]+",a.text)[0])
        recipe['recipe'+str(num)]["title"] = ti[0].text
    my_driver.close()
```
